## 데이터 읽기

```python
import pandas as pd # csv 파일 읽기 위해서 import 한다.

data = pd.read_csv('data/news_keywords.csv', encoding='CP949', header=0)
# pd.read_csv() : pandas에서 csv 파일 읽어들이는 함수
# encoding=utf-8 : 한글은 보통 utf-8을 많이 사용한다.
# 혹시라도, utf-8 codec을 decode 할수 없다고 에러메세지가 뜬다면 encoding='CP949'를 시도한다.
# header=0 하면 0번째 row가 column의 이름이 된다.
# header=None : column의 이름을 지정하지 않겠다. None으로 하게 되면 임의로 숫자가 순서대로 배정된다.
```

```python
data
```

```python
data.values
```

```python
documents = data['Keyword']
```

```python
print(len(documents))
```

```python
total=''
for document in documents:
    total = total + ' ' + document

words=total.split(',')
print(words[0:10])
print(len(words))
```

```python
from collections import Counter
counts=Counter(words)
print(counts)
```

```python
# 단어들에 번호를 매겨 그 번화와 단어를 dictionary로 저장. {단어: index}

word2idx={word.encode("utf8").decode("utf8"): i for i, word in enumerate(counts, 1)}
```

```python
word2idx
```

```python
idx2word={i: word for i, word in enumerate(counts)}
idx2word
```

```python
from sklearn.feature_extraction.text import CountVectorizer
V = len(word2idx)
N = len(documents)
tf = CountVectorizer()
tf.fit_transform(documents)
```

```python
tf.fit_transform(documents)[0:1].toarray() # sample
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf=TfidfVectorizer(max_features=100, max_df=0.95, min_df=0)

A_tfidf_sp=tfidf.fit_transform(documents)
```

```python
tfidf_dict=tfidf.get_feature_names()
print(tfidf_dict)
```

```python
data_array=A_tfidf_sp.toarray()
data=pd.DataFrame(data_array, columns=tfidf_dict)
data.shape
```

```python
from sklearn.manifold import TSNE

tsne=TSNE(n_components=2, n_iter=10000, verbose=1)
```
