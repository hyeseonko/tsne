```python
import pandas as pd
data = pd.read_csv('data/news_keywords.csv', encoding='CP949')
data
```

```python
documents = data['Keyword']
print(len(documents))	# output = 310
total = ''		# 모든 글자를 포함할 string
for document in documents:	
    total = total + ' ' + document

words=total.split(',')	# 쉼표를 구분점으로 하여 words에 저장한다.
print(words[0:10])		# output = [‘신작’, ‘게이머들’, ‘열풍’, …]
print(len(words))	
```

```python
from collections import Counter
counts = Counter(words)
print(counts)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100, max_df=0.95, min_df=0)
doc_tfidf = tfidf.fit_transform(documents)	# size = (310,100)    
tfidf_dict = tfidf.get_feature_names()	# tf-idf dictionary
print(tfidf_dict)	# TF-IDF 로 추출된 100개의 단어
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, n_iter=10000, verbose=1)

data_array = doc_tfidf.toarray() # list를 array로 바꾼다.
Z = tsne.fit_transform(data_array.T)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (10,10)

plt.scatter(Z[:,0], Z[:,1])
for i in range(len(tfidf_dict)):
    plt.annotate(s=tfidf_dict[i], xy=(Z[i,0], Z[i,1]))
plt.draw()
```
