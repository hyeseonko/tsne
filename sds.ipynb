```python
mask=np.zeros(lfw_people.target.shape, dtype=np.bool)
```

```python
mask.shape
```

```python
print(mask)
```

```python
mask2=np.zeros(lfw_people.target.shape, dtype=np.bool)
print(mask2)
```

```python
for target in np.unique(lfw_people.target):
    mask[np.where(lfw_people.target==target)[0][:50]] = 1

print(np.unique(lfw_people.target))
print(lfw_people.target)
print(mask[:500])

X_people=lfw_people.data[mask]
y_people=lfw_people.data[mask]
```

```python
print(lfw_people.data)
print(lfw_people.data[mask2])
```

```python
X_people=lfw_people.data[mask]
y_people=lfw_people.target[mask]
X_people=X_people/255
print(X_people)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test=train_test_split(X_people, y_people, stratify=y_people, random_state=0)
```

```python
X_test
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Accuracy: {:.2f}".format(knn.score(X_test, y_test)))
```

```python
print(len(lfw_people.target_names))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca=pca.transform(X_train)
X_test_pca=pca.transform(X_test)
print(X_train_pca.shape)
print(X_test_pca.shape)
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Accuracy: {:.2f}".format(knn.score(X_test_pca, y_test)))
```

```python
%matplotlib inline
import mglearn
mglearn.plots.plot_pca_whitening()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(2,5, figsize=(10,5), subplot_kw={'xticks':(), 'yticks':()})
#fig, axes = plt.subplots(2,5, figsize=(15,12))

image_shape=lfw_people.images[0].shape
print(image_shape)

for i, (component, ax) in enumerate(zip(pca.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape), cmap='Greys')
    ax.set_title("PCA {}".format((i+1)))
```

```python
mglearn.plots.plot_pca_faces(X_train, X_test, image_shape)
```

```python
from sklearn.decomposition import NMF
nmf=NMF(n_components=100, random_state=0).fit(X_train)

X_train_nmf=nmf.transform(X_train)
X_test_nmf=nmf.transform(X_test)

print(X_train_nmf.shape)
print(X_test_nmf.shape)

fig,axes=plt.subplots(2,5, figsize=(10,5), subplot_kw={'xticks':(), 'yticks':()})
for i, (component, ax) in enumerate(zip(nmf.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape), cmap='Greys')
    ax.set_title("NMF {}".format(i+1))
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_nmf, y_train)
print("Accuracy: {:.2f}".format(knn.score(X_test_nmf, y_test)))
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=100, random_state=0).fit(X_train)

X_train_kmeans=kmeans.transform(X_train)
X_test_kmeans=kmeans.transform(X_test)
```

```python
print(X_train_kmeans.shape)	 	# output = (1547, 100)
print(X_test_kmeans.shape) 	 	# output = (516, 100)
```

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5), 	subplot_kw={'xticks':(), 'yticks':()})

for i, (component, ax) in enumerate(zip(kmeans.components_, axes.ravel())):
    ax.imshow(component.reshape(image_shape), cmap='Greys')
    ax.set_title("K-means {}".format((i+1)))
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_kmeans, y_train)
print("Accuracy: {:.2f}".format(knn.score(X_test_kmeans, y_test)))
```

```python
fig, axes=plt.subplots(3, 5, figsize=(10,5), subplot_kw={'xticks': (), 'yticks':()})
for ax, fe_pca, fe_nmf, fe_kmeans in zip(axes.T, pca.components_, nmf.components_, kmeans.cluster_centers_):
    ax[0].imshow(fe_pca.reshape(image_shape))
    ax[1].imshow(fe_nmf.reshape(image_shape))
    ax[2].imshow(fe_kmeans.reshape(image_shape))

axes[0,0].set_ylabel("PCA")
axes[1,0].set_ylabel("NMF")
axes[2,0].set_ylabel("K-means")
```

```python
X_reconstructed_pca=pca.inverse_transform(pca.transform(X_test))
X_reconstructed_nmf=np.dot(nmf.transform(X_test), nmf.components_)
X_reconstructed_kmeans=kmeans.cluster_centers_[kmeans.predict(X_test)]
```

```python
fig, axes=plt.subplots(4, 5, figsize=(5,5), subplot_kw={'xticks': (), 'yticks':()})
for ax, ori, ir_pca, ir_nmf, ir_kmeans in zip(axes.T, X_test, X_reconstructed_pca, X_reconstructed_nmf, X_reconstructed_kmeans):
    ax[0].imshow(ori.reshape(image_shape))
    ax[1].imshow(ir_pca.reshape(image_shape))
    ax[2].imshow(ir_nmf.reshape(image_shape))
    ax[3].imshow(ir_kmeans.reshape(image_shape))

axes[0,0].set_ylabel("Original")
axes[1,0].set_ylabel("PCA")
axes[2,0].set_ylabel("NMF")
axes[3,0].set_ylabel("K-means")
```

```python
for i in range(50, 510, 50):
    pca=PCA(n_components=i, whiten=True, random_state=0).fit(X_train)
    X_train_pca = pca.transform(X_train)	# apply PCA to X_train 
    X_test_pca = pca.transform(X_test) 	# apply PCA to X_test 
    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train_pca, y_train)
    print("Component: {}".format(i), "Accuracy: {:.2f}".format(knn.score(X_test_pca, y_test)))
```
